```python
import pandas as pd
from sklearn import linear_model
```

```python
df= pd.read_csv("C:\\Users\\archa\\data_science\\git\\my_datascience\\linear_regression\\carprices.csv")
```

```python
df
```

```python
dummies= pd.get_dummies(df["Car Model"],dtype=int)
dummies
                        
```

```python
merge=pd.concat([df,dummies],axis='columns')
merge
```

```python
p_data=merge.drop(['Car Model','Mercedez Benz C class'],axis='columns')
p_data
            
```

```python
X=p_data.drop(['Sell Price($)'],axis='columns')
X
```

```python
y=p_data['Sell Price($)']
y
```

```python
reg= linear_model.LinearRegression()
model=reg.fit(X.values,y)
```

```python
model.predict([[60000,7,0,0]])
```

```python
model.score(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```
